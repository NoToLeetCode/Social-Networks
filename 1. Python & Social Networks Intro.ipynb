**SIMPLE SOCIAL NETWORK USING DICTIONARY**

    Each person is a node, list shows connections

```python
# Create a simple social network using a dictionary
# Each key is a person (node)
# Each value is a list of that person's connections (edges)
network = {
    "Alice": ["Bob", "Charlie"],
    "Bob": ["Alice", "David"],
    "Charlie": ["Alice"],
    "David": ["Bob"]
}

# Degree = number of connections a node has

# Define a function to calculate degree
def degree(node):
    return len(network[node])

# Print the degree of Alice
print(degree("Alice"))
```

**VISUALIZING SIMPLE NETWORK**

    NetworkX creates and draws graphs

```python
# Import libraries for network analysis and plotting
import networkx as nx
import matplotlib.pyplot as plt

# Create an empty graph
G = nx.Graph()

# Add edges from the dictionary-based network
for node in network:
    for neighbor in network[node]:
        G.add_edge(node, neighbor)

# Draw the network with node labels
nx.draw(G, with_labels=True)

# Display the plot
plt.show()
```

**CONTROL STRUCTURES: LOOPS & CONDITIONS**

**FOR LOOP**

    Repeats an action for every item in a list

```python
students = ["Alice", "Bob", "Charlie"]

for name in students:
    print(name)
```

**FOR LOOP WITH SOCIAL NETWORK DATA**

```python
network = {
    "Alice": ["Bob", "Charlie"],
    "Bob": ["Alice"],
    "Charlie": ["Alice", "David"],
    "David": ["Charlie"]
}

for person in network:
    print(person, "has", len(network[person]), "friends")
```

WHILE LOOP

    Runs as long as condition is True

```python
count = 0

while count < 5:
    print("Count is:", count)
    count += 1
```

**IF - ELIF - ELSE STATEMENT**

```python
score = 75

if score >= 90:
    print("Grade A")
elif score >= 70:
    print("Grade B")
elif score >= 50:
    print("Grade C")
else:
    print("Fail")
```

**NETWORK-BASED CONDITIONAL**

    Find popular users (2 or more friends)

```python
network = {
    "Alice": ["Bob", "Charlie", "David"],
    "Bob": ["Alice"],
    "Charlie": ["Alice"],
    "David": ["Alice", "Eva"],
    "Eva": ["David"]
}

for person in network:
    if len(network[person]) >= 2:
        print(person, "is popular")
```

**RANDOM SOCIAL NETWORK**
    
    Automatically generate network

```python
RG = nx.gnm_random_graph(20, 40)
nx.draw(RG)
plt.show()
```

**CUSTOMIZED NETWORK VISUALIZATION**

```python
pos = nx.spring_layout(RG)
nx.draw(RG, pos, node_size=500, node_color="skyblue", edge_color="gray")
plt.show()
```

**DEGREE DISTRIBUTION**


    How many connections nodes have

```python
degrees = [d for n, d in RG.degree()]
plt.hist(degrees)
plt.title("Degree Distribution")
plt.xlabel("Degree")
plt.ylabel("Frequency")
plt.show()
```

**CENTRALITY MEASURES**

    Importance of nodes

```python
# Calculate Degree Centrality
# Measures how many direct connections each node has
degree_centrality = nx.degree_centrality(RG)

# Calculate Betweenness Centrality
# Measures how often a node lies on shortest paths between other nodes
betweenness = nx.betweenness_centrality(RG)

# Print first 5 nodes and their degree centrality values
print(list(degree_centrality.items())[:5])

# Print first 5 nodes and their betweenness centrality values
print(list(betweenness.items())[:5])
```

**CENTRALITY PLOTS**

```python
# Plot Degree Centrality values
# Shows how popular each node is

plt.plot(list(degree_centrality.values()))   # Convert values to list and plot
plt.title("Degree Centrality")               # Add title
plt.xlabel("Node Index")                     # X-axis label
plt.ylabel("Centrality Value")               # Y-axis label
plt.show()                                   # Display plot


# Plot Betweenness Centrality values
# Shows how important nodes are as bridges

plt.plot(list(betweenness.values()))         # Convert values to list and plot
plt.title("Betweenness Centrality")          # Add title
plt.xlabel("Node Index")                     # X-axis label
plt.ylabel("Centrality Value")               # Y-axis label
plt.show()                                   # Display plot
```
